```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dulieuxettuyendaihoc.csv', index_col="STT")
df.head(10)
```

```python
df.tail(10)
```

```python
#Thống kê dữ liệu thiếu cho cột dân tộc và hiệu chỉnh dữ liệu thiếu như sau: Mặc định thiếu  thì điền giá trị 0
df['DT'] = df['DT'].fillna(0)
df
```

```python
df['T1'] = df['T1'].fillna(df['T1'].mean())
df
```

```python
#Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại.
cot_diem = ['L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1', 'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2', 'L3', 'H3', 'S3', 'V3', 'X3', 'D3', 'N3', 'L4', 'H4', 'S4', 'V4', 'X4', 'D4', 'N4', 'L5', 'H5', 'S5', 'V5', 'X5', 'D5', 'N5', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6', 'DH1', 'DH2', 'DH3']
for col in cot_diem:
    df[col] = df[col].fillna(df[col].mean())
df
```

```python
#Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12.
df['TBM1'] = (
    df['T1']*2 + df['L1'] + df['H1'] + df['S1'] +
    df['V1']*2 + df['X1'] + df['D1'] + df['N1']
) / 10

df['TBM2'] = (
    df['T2']*2 + df['L2'] + df['H2'] + df['S2'] +
    df['V2']*2 + df['X2'] + df['D2'] + df['N2']
) / 10

df['TBM3'] = (
    df['T6']*2 + df['L6'] + df['H6'] + df['S6'] +
    df['V6']*2 + df['X6'] + df['D6'] + df['N6']
) / 10

df[['TBM1', 'TBM2', 'TBM3']].head()
```

```python
#Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp 10, 11, 12
def xep_loai(tbm):
    if tbm < 5.0:
        return 'Y'
    elif tbm < 6.5:
        return 'TB'
    elif tbm < 8.0:
        return 'K'
    elif tbm < 9.0:
        return 'G'
    else:
        return 'XS'
df['XL1'] = df['TBM1'].apply(xep_loai)
df['XL2'] = df['TBM2'].apply(xep_loai)
df['XL3'] = df['TBM3'].apply(xep_loai)
df[['TBM1', 'XL1', 'TBM2', 'XL2', 'TBM3', 'XL3']].head()
```

```python
#Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp 10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương pháp Min-Max Normalization
df['US_TBM1'] = (df['TBM1'] / 10) * 4
df['US_TBM2'] = (df['TBM2'] / 10) * 4
df['US_TBM3'] = (df['TBM3'] / 10) * 4

df[['TBM1', 'US_TBM1', 'TBM2', 'US_TBM2', 'TBM3', 'US_TBM3']].head()
```

```python
#Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị “1”) và rớt ( giá trị “0”) vào các khối dựa trên điểm DH1, DH2 và DH3
def xet_tuyen(row):
    if row['KT'] in ['A', 'A1']:
        diem = (row['DH1']*2 + row['DH2'] + row['DH3']) / 4
    elif row['KT'] == 'B':
        diem = (row['DH1'] + row['DH2']*2 + row['DH3']) / 4
    else:
        diem = (row['DH1'] + row['DH2'] + row['DH3']) / 3

    return 1 if diem >= 5.0 else 0
df['KQXT'] = df.apply(xet_tuyen, axis=1)
df[['KT', 'DH1', 'DH2', 'DH3', 'KQXT']].head()
```

```python
#Lưu trữ dữ liệu xuống ổ đĩa thành file processed_dulieuxettuyendaihoc.csv
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
```
