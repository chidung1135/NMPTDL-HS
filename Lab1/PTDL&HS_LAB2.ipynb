2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau:

a. Từ 0 đến dưới 5 = kém (ký hiệu “k”)

b. Từ 5 đến dưới 7 = trung bình (ký hiệu “tb”)

c. Từ 7 đến dưới 8 = khá (ký hiệu “k”)

d. Từ 8 trở lên = giỏi (ký hiệu “g”)


```python
def phanloai_t1(x):
    if x < 5:
        return "k"
    elif x < 7:
        return "tb"
    elif x < 8:
        return "kh"
    else:
        return "g"
df["phanlopt1"] = df["T1"].apply(phanloai_t1)
df[["T1", "phanlopt1"]].head()
```

3. Lập bảng tần số cho biến phanloait1


```python
tan_so_phanlopt1 = df["phanlopt1"].value_counts()
tan_so_phanlopt1
```

4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanlopt1


```python
df[["T1", "phanlopt1"]].head()
for loai in df["phanlopt1"].unique():
    subset = df[df["phanlopt1"] == loai]
    plt.plot(subset.index, subset["T1"], label=loai)
plt.xlabel("T1")
plt.ylabel("Count")
```

```python
multi_line_data = (
    df.groupby(["T1", "phanlopt1"])
      .size()
      .reset_index(name="Count")
)
for nhom in multi_line_data["phanlopt1"].unique():
    data_nhom = multi_line_data[multi_line_data["phanlopt1"] == nhom]
    plt.plot(
        data_nhom["T1"],
        data_nhom["Count"],
        label=nhom
    )
plt.xlabel("T1")
plt.ylabel("Count")
plt.legend(title="Phanlopt1")


```

5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1


```python
drop_data = df.groupby(["T1", "phanlopt1"]).size().reset_index(name="Count")
for nhom in drop_data["phanlopt1"].unique():
    data_nhom = drop_data[drop_data["phanlopt1"] == nhom]

    plt.vlines(
        x=data_nhom["T1"],
        ymin=0,
        ymax=data_nhom["Count"],#
        linewidth=2
    )
    plt.scatter(
        data_nhom["T1"],
        data_nhom["Count"],
        label=nhom
    )

plt.xlabel("T1")
plt.ylabel("Count")
plt.legend(title="Phanlopt1")
```
